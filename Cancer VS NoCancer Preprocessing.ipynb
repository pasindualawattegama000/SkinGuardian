```python
import pandas as pd

# Load labels from Excel file
Directory = r"C:\Users\pasin\Documents\IIT\FYP\Skin Cancer Detection cnn\SkinCancer vs Diesase CSV\skinCancer & disease.csv"
labels_df = pd.read_csv(Directory)
image_ids = labels_df['image_id'].tolist()
cancer_types = labels_df['label'].tolist()
```

```python
# Preprocess the image filenames
image_folder = 'C:/Users/pasin/Documents/IIT/FYP/Skin Cancer Detection cnn/All cancer and non cancer images/'

image_paths = [image_folder + image_id + '.jpg' for image_id in image_ids]
```

```python
# Load and preprocess the images
import cv2
import numpy as np
```

```python
# Define the image size
image_size = (200, 200)

# Create empty arrays for X and Y variables
X = []
Y = []
```

```python
# Iterate over the image paths
for image_path, cancer_type in zip(image_paths, cancer_types):
    # Load and resize the image
    image = cv2.imread(image_path)
    image = cv2.resize(image, image_size)
    # Normalize pixel values (e.g., scale to [0, 1])
    image = image.astype(np.float32) / 255.0
    # Append the preprocessed image to X
    X.append(image)
    # Append the cancer type to Y
    Y.append(cancer_type)
```
